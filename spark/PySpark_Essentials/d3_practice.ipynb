```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("D3").getOrCreate()
spark
```

```python
fifa = spark.read.csv('Datasets/fifa19.csv', inferSchema=True, header=True)
```

```python
fifa.limit(5).toPandas()
```

```python
fifa.createOrReplaceTempView('fifaView')
```

#### How many players are there in each position?

```python
df1 = spark.sql("SELECT Position, count(*) AS Count FROM fifaView GROUP BY Position ORDER BY Count DESC")
df1.limit(10).toPandas()
```

```python
# Writing this df1 to json file
df1.toPandas().to_csv("data/fifa_position_count.csv")
```
